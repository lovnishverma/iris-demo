# Encoding

## Label Encoding

* Suitable for ordinal (ranked) categories

**PPT:** https://github.com/lovnishverma/Slidespptspdfs/blob/main/Data%20Science/Python%20Basics%20Refresher.pdf

```python
df['label'].value_counts()
```

```python
le = LabelEncoder()

df['label'] = le.fit_transform(df['label'])
```

```python
df.head()
```

```python
df['label'].value_counts()
```

```python
# decode

df['label'] = le.inverse_transform(df['label'])
```
